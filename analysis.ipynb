### MISSING VALUES

```python
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100
missing_summary = pd.DataFrame({
    'Missing Count': missing_data,
    'Missing %': missing_percent.round(2)
})
display(missing_summary[missing_summary['Missing Count'] > 0])
```

```python
print("The only columns which has missing values is : ")
df[["ai_keywords", "ai_skills"]].isna().sum()
```

### ALWAYS MAKE COPY OF YOUR DATASET BEFORE MOBIFYING

```python
df_copy = df.copy()
```

### FILLING MISSING VALUES

```python
# fill the missing values with 'Unknown'
df_copy[["ai_keywords", "ai_skills"]] = df_copy[["ai_keywords", "ai_skills"]].replace(np.NaN, 'Unknown')
```
